```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Create mock data for 100 patients
num_patients = 100
patient_ids = ["Patient_" + str(i + 1) for i in range(num_patients)]

# Generate realistic systolic and diastolic blood pressure readings
systolic = np.random.normal(loc=125, scale=15, size=num_patients)
systolic = np.round(systolic).astype(int)

diastolic = np.random.normal(loc=80, scale=10, size=num_patients)
diastolic = np.round(diastolic).astype(int)

# Create DataFrame
df_bp = pd.DataFrame({
    "PatientID": patient_ids,
    "Systolic": systolic,
    "Diastolic": diastolic
})

# Print first rows
print(df_bp.head())

# Create blood pressure histograms
plt.figure(figsize=(12, 6))

plt.hist(df_bp["Systolic"], bins=15, alpha=0.7, label="Systolic")
plt.hist(df_bp["Diastolic"], bins=15, alpha=0.7, label="Diastolic")

plt.xlabel("Blood Pressure (mmHg)")
plt.ylabel("Frequency")
plt.title("Distribution of Blood Pressure Readings for 100 Patients")
plt.legend()
plt.tight_layout()

plt.show()
```
